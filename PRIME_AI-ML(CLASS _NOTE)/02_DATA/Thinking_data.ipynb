#### Problem in this data

Data is inconsistent (rating) --> mixed type(int, float, string)

missing data (age)

Duplicate data

# ASSIGNMENT-1

clean and structure the data

```python
# clean and structure the data

def clean_data(data):
    text_to_num = {"one" : 1, "two" : 2, "three" : 3, "four" : 4, "five" : 5}
    cleaned_data = []
    unique_users = set()

    for user in data:
        # clean rating
        raw_rating = str(user["rating"]).strip().lower() # strip remove all the space
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating
        # print(raw_rating)

        # Handle missing value
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None
        # print(raw_age)

        # Deduplication (Clean Duplicate Data)
        if(user["name"].strip() in unique_users):
            continue
        
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
```

```python
data = clean_data(data)
```

# ASSIGNMENT - 2

Getting meaningful insights from data
1) What average rating
2) what % of users gave poor rating(< 3)?

```python
def get_insights(data):
    # average rating
    total_rating = 0
    for user in data:
        total_rating += float(user["rating"])
    print(f"average rating is {total_rating / len(data)}")

    # % of user gave poor rating(<3)
    poor_rating = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_rating+=1
    print(f"% of user with poor rating is {poor_rating / len(data) * 100}")
```

```python
get_insights(data)
```

# ASSIGNMENT - 3
Build a product recommendation feature(Rule base not ml)q
1) If user rating >= 4, recommend same brand product
2) if user rating < 4, recommend different brand product

```python
# Build recommendation function
def recommend_feature(data):
    recommendation = []
    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]
        if(float(user["rating"] ) >= 4):
            curr_recomm["brand"] = "Apple"
        else:
            curr_recomm["brand"] = "Samsung"
    
        recommendation.append(curr_recomm)
    return recommendation
```

```python
recommend_feature(data)
```
