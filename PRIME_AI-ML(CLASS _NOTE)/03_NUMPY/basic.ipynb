Python-list vs numpy array

1) Numoy array are faster (optimized in c)
2) Numpy stores data in continuous memory block
3) Used less memory (optimized storage)
4) Build in mathematical function
5) Support direct vectorized operations

```python
import time
```

```python
# compare execution performance
size = 10000000
# python list
py_lst = list(range(size))
srt_time = time.time()
sq_lst = [x**2 for x in py_lst]
end_time = time.time()
print(f"Python list take {end_time - srt_time} time")

# numpy array
np_arr = np.array(py_lst)
srt_time = time.time()
sq_arr = np_arr**2
end_time = time.time()
print(f"Numpy array take {end_time - srt_time} time")
```

```python
# memory

import sys

print(f"python list syze = {sys.getsizeof(py_lst) * len(py_lst)} bytes")

print(f"Numpy array syze = {np_arr.nbytes} bytes")
```

So wht numpy is faster
1) Homogenous data types and contigunous memory
2) vector operation and C implementations
3) Reduces overhead
