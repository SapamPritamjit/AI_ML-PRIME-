```python
import numpy as np
```

* integer -> int8, int16, int32, int64, int128

```python
arr = np.array([1, 2, 3, 4, 5])
print(arr, arr.dtype)
arr = np.array([1, 2, 3, 4, 5], dtype="int16")
print(arr, arr.dtype)
```

* floating point numbers -> float8, float16, float32, float64

```python
arr = np.array([1, 2, 3.1, 4, 5])
print(arr, arr.dtype)
arr = np.array([1, 2, 3.1, 4, 5], dtype="float32")
print(arr, arr.dtype)
```

* Boolean : bool

* complex numbers --> complex64, complex128

```python
arr1 = np.array([3 + 5j]) # complex128 -> float64 + float64
print(arr1, arr1.dtype)
```

```python
arr2 = np.array([2 + 6j])
print(arr1 + arr2)
```

## NOT RECOMMEND TO USE IN NUMPY

* String : S (byte-str) & U (unicode - str)

* Object : generic python objects -object
