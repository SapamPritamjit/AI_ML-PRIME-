* Fancy and Boolean indexing

1) it can excess multiple index at same time
2) As for fancy indexing we can simply write the index number in a list
3) As for boolean it depending upon a particular condition -- if it true it print the value

```python
# Fancy indexing

arr = np.array([56, 67, 435, 22, 34])
idx = [1, 2, 3] # index that i want to excess
print(arr[idx])
```

```python
# Boolean indexing

arr = np.array([1, 2, 3, 4, 5])

print(arr[arr % 2 == 0])
print(arr[arr > 2])
```

* Slicing

```python
arr = np.array([1, 2, 3, 4, 5, 6, 7])

print(arr[1:5:1]) # start : stop : step
print(arr[:])
```

* Copy vs View in slicing

```python
py_lst = [1, 2, 3, 4, 5]
sub_lst = py_lst[1:3]
print(sub_lst)

sub_lst[0] =  200
print(sub_lst)
print(py_lst)
```

```python
np_arr = np.array([1, 2, 3, 4, 5])
sub_arr = np_arr[1:3]
print(sub_arr)

sub_arr[0] = 10
print(sub_arr)
print(np_arr)
```
