* Using build in function

```python
# np.arange() -> create an array with a range of values
arr = np.arange(1, 10, 1) # start, end, step
print(arr)
```

```python
# np.zeros() -> creating an array filled with zeroes

arr = np.zeros((2, 3))  # 2 row 3 column
print(arr)
```

```python
# np.ones() -> creates an array filled with ones

arr = np.ones((2, 3), dtype="int64")
print(arr)
```

```python
# np.full() -> create an array filled with the value we give

arr = np.full((2, 3), 8)
print(arr)
```

```python
# np.eye() -> create an array and filled the diagonal with "1" and "0" with remaining

arr = np.eye(3)
print(arr)
```

```python
# np.linspace() -> it create evenly space over a specific range

arr = np.linspace(1, 100, 4) # start, end, total number of value
print(arr)
```
