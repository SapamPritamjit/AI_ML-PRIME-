# Histogram :-
A histogram shows the distribution of numerical data by:
* Dividing values into bins(ranges)
* Counting how many values fall into each bin (frequency)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
score = np.random.normal(70, 10, 100) # normal distribution
```

```python
plt.hist(score, edgecolor = "black", bins = 5) # default bins
plt.ylabel("Number / frequency of students")
plt.xlabel("mark obtained")
plt.show()
```

```python
bins =[0, 30, 70, 90, 100]

plt.hist(score, edgecolor = "black", bins = bins) # custom bins
plt.ylabel("Number / frequency of students")
plt.xlabel("mark obtained")
plt.show()
```

```python
marks = [
    25, 38, 42, 55, 61, 67, 72, 78, 83, 88,
    91, 95, 47, 53, 59, 64, 69, 74, 79, 85,
    92, 97, 33, 45, 58, 62, 71, 76, 81, 89
]
```

```python
bins =[0, 30, 70, 90, 100]

plt.hist(marks, edgecolor = "black", bins = bins) # custom bins
plt.ylabel("Number of students")
plt.xlabel("mark obtained")
plt.title("distribution of marks for students")
plt.grid(True)
plt.show()
```

* MULTIPLE DATASET

```python
legit_transactions = [
    2.99, 5.49, 8.99, 12.50, 14.99, 19.99, 23.45, 29.99, 34.99, 39.50,
    45.00, 49.99, 55.25, 60.00, 75.99, 89.99, 120.50, 150.00, 199.99,
    249.99, 300.75, 450.00, 600.00, 850.00, 1200.00
]

fraud_transactions = [
    50, 100, 150, 200, 300, 500, 500, 750, 1000, 1000,
    1200, 1500, 1500, 2000, 2500, 3000, 3000
]
```

```python
plt.hist(fraud_transactions, color = "red", edgecolor = "black", label = "Fraud", alpha = 0.5, bins = 20)
plt.hist(legit_transactions, color="green", edgecolor = "black", label = "Legit", alpha = 0.6, bins = 20)
plt.xlabel('Transaction Amount ($)')
plt.ylabel('Frequency')
plt.title('Transaction Amount Distribution: Legitimate vs Fraudulent')
plt.legend()
plt.grid(True)
```

* axvline :- It is used to draw a vertical reference line on a plot, common to mark cutoffs, averages, pass marks or thresholds


```python
# axv line

# custom bins
bins = [0, 30, 70, 90, 100]
plt.hist(
    marks,
    bins=bins,
    edgecolor="black",
    color="lightblue"
)

# line at passing marks - 33
plt.axvline(
    x=33,
    color="red",
    linestyle="--",
    linewidth=2,
    label="Pass Mark (33)"
)

plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks of 30 Students")
plt.grid(True)
```
