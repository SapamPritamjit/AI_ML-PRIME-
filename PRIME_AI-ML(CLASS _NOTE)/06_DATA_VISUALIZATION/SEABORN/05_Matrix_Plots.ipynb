# Matrix Plots :-
* Heatmaps - Plot rectangular data as color encoded matrix

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flight_df = sns.load_dataset("flights")
flight_df.head()
```

```python
flight_pivot = flight_df.pivot(index = "month", columns= "year", values="passengers")
```

```python
sns.heatmap(
    flight_pivot,
    cmap="coolwarm" # optional
)
```

```python
sns.heatmap(
    flight_pivot,
    cmap="coolwarm", # optional
    annot=True, # add number
    fmt="d" # integer format
)

plt.title("Passenger Heatmap")
plt.tight_layout()
```
