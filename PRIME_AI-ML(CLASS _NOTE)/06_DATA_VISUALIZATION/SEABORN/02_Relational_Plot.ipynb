# Relational Plot :-
* Scatter Plots
* Line Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
# Tip and Total_Bill

sns.scatterplot(
    data = tips_df,
    x = "total_bill",
    y = "tip",
    hue = "time"
)
```

```python
# Line Plots

sns.lineplot(
    data = tips_df,
    x = "total_bill",
    y = "tip",
    hue = "time"
)
```

```python
flight_df = sns.load_dataset("flights")
flight_df.head()
```

```python
sns.lineplot(
    data = flight_df,
    x = "year",
    y = "passengers"
)
```
