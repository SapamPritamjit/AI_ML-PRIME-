# Categorical Plots :-
* Bar Plots
* Box Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
sns.barplot(
    data = tips_df,
    x = "day",
    y = "tip",
    hue = "sex"
)
```

```python
sns.boxplot(
    data = tips_df,
    x = "day",
    y = "tip",
    # hue = "sex"
)
```
