```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
fig, ax = plt.subplots()
sns.lineplot(
    data = tips_df,
    x = "day",
    y = "total_bill",
    hue = "sex",
    ax = ax,
    marker = "o"
)
ax.set_title("avg bill value for each days")
```

```python
fig, ax = plt.subplots()
sns.lineplot(
    data = tips_df,
    x = "day",
    y = "total_bill",
    hue = "sex",
    ax = ax,
    marker = "o",
    errorbar=None
)
ax.set_title("avg bill value for each days")
```
