# Pie Charts :-
A pie charts is a circular chart divided into slices to show relative proportion of a whole.
* Each slice = a category
* Size of slice = value of that category relation to total

```python
import matplotlib.pyplot as plt
```

```python
expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
```

```python
plt.pie(amounts, labels=expenses)
```

* autopct = " %1.1f%% " - shows percentage of each slice.
* colors - add custom colors
* wedgeprops - define (edgecolor, linewidth, linestyle, alpla, fill..)
* explode - highlight a slice by pulling it out
* startrangle - rotate the charts
* shadow - adds 3D effect or makes the chart circular

```python
plt.pie(amounts, labels=expenses, autopct="%1.1f%%")
plt.show()
```

```python
# wedgeprops

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
        wedgeprops={
            "edgecolor" : "black",
            "linewidth" : 1,
            "linestyle" : "--",
            "fill" : False, # remove all the color that filled the slice
            "alpha" : 0.7 # color darknedd ratio
        }
        )
plt.show()
```

```python
# explode
explode = [0, 0, 0, 0.2, 0] # give number as 0 if we dont want to pilled it out that slice

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
        wedgeprops={
            "edgecolor" : "black",
            "linewidth" : 1,
            "linestyle" : "--",
        },
        explode=explode
        )
plt.show()
```

```python
explode = [0, 0.1, 0, 0.2, 0]

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
        wedgeprops={
            "edgecolor" : "black",
            "linewidth" : 1,
            "linestyle" : "--",
        },
        explode=explode
        )
plt.show()
```

```python
# Startangle

explode = [0, 0, 0, 0.2, 0]

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
        wedgeprops={
            "edgecolor" : "black",
            "linewidth" : 1,
            "linestyle" : "--",
        },
        explode=explode,
        startangle=90
        )
plt.show()
```

```python
# shadow

# Startangle

explode = [0, 0, 0, 0.2, 0]

plt.pie(amounts, labels=expenses, autopct="%1.1f%%",
        wedgeprops={
            "edgecolor" : "black",
            "linewidth" : 1,
            "linestyle" : "--",
        },
        explode=explode,
        startangle=90,
        shadow=True
        )
plt.show()
```
