```python
import matplotlib.pyplot as plt
```

```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M

# Non oscar winning movies

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M
```

```python
plt.plot(oscar_movies, oscar_revenue)
plt.title("Oscar Movie Revenue")
plt.xlabel("Oscar Movies")
plt.ylabel("Revenue")
plt.tight_layout()
```

* Adding Color

```python
plt.plot(oscar_movies, oscar_revenue, color = "red")
plt.title("Oscar Movie Revenue")
plt.xlabel("Oscar Movies")
plt.ylabel("Revenue")
plt.tight_layout()
```

* Adding Style

```python
plt.plot(oscar_movies, oscar_revenue, linestyle = "--", linewidth = 3)
plt.title("Oscar Movie Revenue")
plt.xlabel("Oscar Movies")
plt.ylabel("Revenue")
plt.tight_layout()
```
