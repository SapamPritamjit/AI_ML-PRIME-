# Box Plot :-
It is a statistical visualization that summerizes the distribution of a dataset using five key numbers :
* Minimum (lowest non-outlier)
* First Quartile(Q1) --> 25 percentile
* Median (Q2) --> 50 percentile
* Third Quartile(Q3) --> 75 percentile
* Maximum (Highest non-outlier)

```python
import matplotlib.pyplot as  plt
```

```python
data = [7, 8, 5, 6, 9, 4, 10, 12, 15]

plt.boxplot(data)
plt.grid(True)
```

* Multiple Dataset

```python
import numpy as np
```

```python
group1 = np.random.normal(50, 10, 100)
group2 = np.random.normal(60, 15, 100)
```

```python
plt.boxplot([group1, group2], tick_labels=["group1", "group2"])
plt.grid(True)
plt.show()
```

* Horizontal Box Plot

```python
plt.boxplot([group1, group2], tick_labels=["group1", "group2"], vert=False)
plt.grid(True)
plt.show()
```

* Show Mean

```python
plt.boxplot([group1, group2], tick_labels=["group1", "group2"], showmeans=True)
plt.grid(True)
plt.show()
```

* Change Whisker Length

```python
plt.boxplot([group1, group2], tick_labels=["group1", "group2"], whis=1.0)
plt.grid(True)
plt.show()
```
