# Stack Plots :-
* It is a type of plot where multiple data series are stacked on top each other.
* Each "layer" shows the contribution of one category and the top line shows the cumulative total.

```python
import matplotlib.pyplot as plt
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
direct = [50, 60, 70, 80, 90, 100, 110]
organic = [30, 40, 50, 55, 60, 70, 80]
social = [20, 25, 30, 35, 40, 50, 60]
```

```python
# for 1st day

plt.pie([direct[0], organic[0], social[0]], labels=["direct", "organic", "social"], autopct="%1.1f")
plt.show()
```

```python
# for whole data

plt.stackplot(days, direct, organic, social, labels=["direct", "organic", "social"])
plt.title("Website Traffic Over a Week")
plt.xlabel("Days")
plt.ylabel("Number of customer")
plt.legend()
```
