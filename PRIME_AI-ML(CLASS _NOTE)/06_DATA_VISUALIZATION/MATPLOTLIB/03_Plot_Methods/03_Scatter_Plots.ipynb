# Scatter Plots :-
* A scatter plot show the relationship between two variables by plotting points on a 2D plane.

```python
import matplotlib.pyplot as plt
```

```python
people = ["Person A", "Person B", "Person C", "Person D", "Person E", "Person F", "Person G", 
          "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
```

```python
plt.scatter(age, blood_pressure)
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

* Adding color

```python
plt.scatter(age, blood_pressure, color = "red")
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

```python
color = ["green" if x < 135 else "red" for x in blood_pressure]

plt.scatter(age, blood_pressure, color = color)
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

* Marker

```python
color = ["green" if x < 135 else "red" for x in blood_pressure]

plt.scatter(age, blood_pressure, color = color, marker = "<")
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

Annotations :- text that points at data and says this is important
* Labeling a specific point
* Explaining a trend or event
* Highlighting outliers
* Adding context (dates, values, names)

plt.annotate("text", xy = (x_point, y_point), xytext = (x_text, y_text))

* xy -> point
* xy_text -> position where we want to show

```python
color = ["green" if x < 135 else "red" for x in blood_pressure]

plt.scatter(age, blood_pressure, color = color, marker = "o")
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()

for i in range(len(people)):
    plt.annotate(people[i], xy=(age[i], blood_pressure[i]), xytext=(age[i], blood_pressure[i]))

plt.xlim(min(age), max(age)+7)
plt.ylim(min(blood_pressure), max(blood_pressure)+10)
```

* Multiple data on a single scatter plot

```python
cities = ["city A", "city B", "city C", "city D", "city E"]

# Winter : Temperature (*C) vs Humidity (%)
winter_temp = [5, 2, 10, 0, 7]
winter_humidity = [80, 75, 65, 85, 70]

# Summer : Temperature (*C) vs Humidity (%)
summer_temp = [25, 30, 28, 35, 27]
summer_humidity = [60, 50, 55, 45, 65]
```

```python
plt.scatter(winter_temp, winter_humidity, label = "winter")
plt.scatter(summer_temp, summer_humidity, label = "summer")
plt.title("Winter VS Summer")
plt.xlabel("Temp (*C)")
plt.ylabel("Humidity (%)")
plt.legend()
```
