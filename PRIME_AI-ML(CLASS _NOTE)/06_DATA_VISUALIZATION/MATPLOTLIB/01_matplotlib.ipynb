```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [4, 5, 6, 7]

plt.plot(x, y) # line plot
```

```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M
```

```python
plt.plot(years, oscar_revenue)
plt.title("Oscar Movies Revenue in each Year")
plt.xlabel("year")
plt.ylabel("Oscar Revenue in $M")
```

```python
# Non oscar winning movies

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M
```

```python
plt.plot(years, oscar_revenue, label = "Oscar Movies")
plt.plot(years, non_oscar_revenue, label = "Non-Oscar Movies")
plt.title("Oscar Movies Revenue VS Non-Oscar Movies Revenue")
plt.xlabel("year")
plt.ylabel("Revenue in $M")
plt.legend() # used to show label
```

```python
plt.plot(years, oscar_revenue, color = "red" ,label = "Oscar Movies")
plt.plot(years, non_oscar_revenue, color = "green", label = "Non-Oscar Movies")
plt.title("Oscar Movies Revenue VS Non-Oscar Movies Revenue")
plt.xlabel("year")
plt.ylabel("Revenue in $M")
plt.legend() # used to show label
```

```python
plt.plot(years, oscar_revenue,color = "brown", linestyle = "-.", linewidth = 5, label = "Oscar Movies")
plt.plot(years, non_oscar_revenue, label = "Non-Oscar Movies")
plt.title("Oscar Movies Revenue VS Non-Oscar Movies Revenue")
plt.xlabel("year")
plt.ylabel("Revenue in $M")
plt.legend() # used to show label
```

```python
plt.plot(years, oscar_revenue, "o--r", label = "Oscar Movies")
plt.plot(years, non_oscar_revenue, "o--g", label = "Non-Oscar Movies")
plt.title("Oscar Movies Revenue VS Non-Oscar Movies Revenue")
plt.xlabel("year")
plt.ylabel("Revenue in $M")
plt.legend() # used to show label
```
