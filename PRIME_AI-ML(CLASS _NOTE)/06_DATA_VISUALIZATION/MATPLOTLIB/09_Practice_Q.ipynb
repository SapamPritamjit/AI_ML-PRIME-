* Compare temperature of cities over a week using a subplot & loop

```python
# Temperature over cities
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

cities = ["New York", "London", "Delhi", "Tokyo"]

temperatures = [
    [22, 23, 21, 24, 25],   # New York
    [18, 19, 17, 20, 21],   # London
    [30, 32, 31, 33, 34],   # Delhi
    [25, 26, 24, 27, 28]    # Tokyo
]
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(2, 2)

city = 0
for row in range(2):
    for col in range(2):
        ax[row][col].plot(days, temperatures[city], marker = "o")
        ax[row][col].set_title(cities[city])
        ax[row][col].grid(True)
        city+=1

fig.suptitle("temperature of cities over the week")
fig.supxlabel("Days")
fig.supylabel("Temperature")
fig.tight_layout()
```

```python
fig, axes = plt.subplots(2, 2)
axes = axes.flatten()

for i, ax in enumerate(axes):
    ax.plot(days, temperatures[i], marker='o')
    ax.set_title(cities[i])
    ax.grid(True)

fig.suptitle("Weekly Temperature by City")
fig.supylabel("Temperature (C)")
fig.supxlabel("Day")

fig.tight_layout()
```
