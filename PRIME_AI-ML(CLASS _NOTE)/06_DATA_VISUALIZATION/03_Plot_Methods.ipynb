```python
import matplotlib.pyplot as plt
```

```python
oscar_movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
oscar_revenue = [1005, 170, 427, 133, 232] # in $M

# Non oscar winning movies

non_oscar_movies = ["Slumdog Millionaire", "Avatar", "Inception", "Hugo", "Lincoln"]
non_oscar_revenue = [378, 2788, 829, 185, 275] # in $M
```

# Line plot

```python
plt.plot(oscar_movies, oscar_revenue)
plt.title("Oscar Movie Revenue")
plt.xlabel("Oscar Movies")
plt.ylabel("Revenue")
plt.tight_layout()
```

# Bar Charts :- 
* A bar plot(or bac chart) is a graph that displays categorical data using rectangular bars
* Each bar's height(or length, in horizontalbars) respresent a value

```python
movies = [
    "The Dark Knight",
    "The Hurt Locker",
    "The King's Speech",
    "The Artist",
    "Argo"
]

years = [2008, 2009, 2010, 2011, 2012]
revenue = [1005, 170, 427, 133, 232] # in $M
```

```python
plt.bar(years, revenue, color = "Brown")
plt.title("Revenue in each year")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")
```

* plt.text() -> (x-axis position, y-axis position, text, alignment) # reqd to pass this 4 value

```python
plt.bar(years, revenue, color = "Green")
plt.title("Revenue in each year")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")

for i in range(len(years)):
    plt.text(years[i], revenue[i] + 10, revenue[i], ha = "center")
```

* xlim  limits the horizontal (x) axis   
     * { (min_value, max_value) }

* ylim  limits the vertical (y) axis     
     * { (min_value, max_value) }

```python
plt.bar(years, revenue, color = "Green")
plt.title("Revenue in each year")
plt.xlabel("Year")
plt.ylabel("Revenue ($M)")

for i in range(len(years)):
    plt.text(years[i], revenue[i] + 20, revenue[i], ha = "center")

plt.ylim(0, max(revenue) + 100) # used to extend the y axis
```

* Compare 2 set of data
* we custom width, width control how thick each bar is. Usually ranges from 0 to 1.[Default value id 0.8]

```python
import numpy as np

plt.title("Oscar VS Non-Oscar movies")
plt.xlabel("Years")
plt.ylabel("Revenue ($M)")

x = np.arange(len(years))
width = 0.4

plt.bar(x - width / 2, oscar_revenue, width, label = "Oscar Revenue")
plt.bar(x + width / 2, non_oscar_revenue, width, label = "Non-Oscar Revenue")
plt.legend()
plt.xticks(x, years) # used to replace the position value with years
```

# Horizontal Bar Charts

```python
plt.barh(oscar_movies, oscar_revenue)
plt.title("Oscar Movies Revenue")
plt.xlabel("Oscar Movies")
plt.ylabel("Revenue in ($M)")
```

# Scatter Plots :-
* A scatter plot show the relationship between two variables by plotting points on a 2D plane.

```python
people = ["Person A", "Person B", "Person C", "Person D", "Person E", "Person F", "Person G", 
          "Person H", "Person I", "Person J"]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
blood_pressure = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
```

```python
plt.scatter(age, blood_pressure)
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

```python
plt.scatter(age, blood_pressure, color = "red")
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```

```python
color = ["green" if x < 135 else "red" for x in blood_pressure]

plt.scatter(age, blood_pressure, color = color)
plt.title("Age VS Blood-Pressure")
plt.xlabel("Age")
plt.ylabel("Blood Pressure")
plt.grid()
```
