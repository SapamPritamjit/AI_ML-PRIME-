* Overfitting: Model fits training data too well but fails to generalize
* Underfitting: Model is too simple to capture underlying patterns

* Overfitting = memorizing
* Underfitting = guessing

# Overfitting (model too complex)
üîç What it means

  * Model learns noise
  * Low training error, high test error
  * High variance

üõ†Ô∏è How to fix overfitting
* Simplify the model

Reduce depth / parameters
* Add regularization
  * L1 (Lasso)
  * L2 (Ridge)
* Use more training data
* Feature selection
  * Remove irrelevant features
* Early stopping
* Dropout (for neural networks)
* Data augmentation

# Model-specific fixes
* Linear / Logistic Regression

  * Overfitting ‚Üí increase regularization
  * Underfitting ‚Üí add features / interactions

* Decision Trees

  * Overfitting ‚Üí limit depth, min samples
  * Underfitting ‚Üí increase depth

* Neural Networks

  * Overfitting ‚Üí dropout, early stopping
  * Underfitting ‚Üí more layers / neurons
