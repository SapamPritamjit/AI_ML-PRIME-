* Feature Engineering :- Transforming, creating or selecting input variables to make our model perform
better.

* Encoding categorical variable (.map(), one hot Encoding)
* Create useful derived Features
* Interaction Features
* Scaling numeric Features (Normalization & Standardization)
* Feature selection

### One hot Encoding
Always use "drop_first=True" to avoid multicollinearity

## Interaction Features :- 
* An interaction feature represents the combined effect of two or more features and is usually created by multiplying them.
### Why interaction features are important
* Linear models cannot learn interactions automatically
* Adding interaction features lets them model non-linear relationships

## Scaling numeric Features :-
* Scaling numeric features ensures all features contribute equally to the model by bringing them to a common scale.
* This prevents features with large values from dominating the model.

* Why scaling is needed

  * Example:
  * Feature	Range
  * Age	0 – 100
  * Income	10,000 – 1,000,000

* Without scaling:
  * Income dominates
  * Age becomes almost useless

## Feature selection :-
* Feature selection is the process of identifying and selecting the most relevant features to improve model performance and reduce complexity.
* Goal: better performance, faster training, simpler models

* Why feature selection is important
  * Reduces overfitting
  * Improves model accuracy
  * Reduces training time
  * Improves interpretability
