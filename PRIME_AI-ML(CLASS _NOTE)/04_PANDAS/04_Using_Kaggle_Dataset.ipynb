```python
import pandas as pd
```

```python
df = pd.read_csv("globalAirQuality.csv")
df.head()
```

```python
df.describe()
```

### selecting data

* select columns

```python
df["city"] # select only one column
df[["city", "aqi", "co"]] # select multiple column
```

* select rows (label & index based) -loc & iloc

```python
# index based

df.iloc[[7, 8]] # iloc -> index location
```

```python
df.iloc[0:6] # with (iloc) it dosont include ending {start - 0, end - 6, step - 1}
```

```python
# label based
df.loc[[2, 4]] # loc -> location
```

```python
df.loc[0:6] # with (loc) it include ending too {start - 0, end - 6, step - 1}
```

* Select Row and Columns

```python
# with loc

df.loc[1, "city"] # row : column
```

```python
# for multiple row and column 

df.loc[0 : 2, ["city", "aqi"]]
```

```python
# with iloc 

# df.iloc[1, "city"] # row : column

# it will give error because "city" is not a numerical value it a string on iloc only numerical value can pass
```

```python
df.iloc[1, 2] # row : column (index 2 which is city)
```

```python
df.iloc[1 : 4, 2]
```

```python
df.iloc[1 : 5, 1 : 5]
```

* Select single scalar value

```python
# with at 

df.at[0, "city"]
```

```python
# with iat

df.iat[0, 2]
```
