* Data Concatenation :- Stacking DataFrames on top of each other (rows) or side by side (columns)

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "id" : [1, 2, 3],
    "name" : ["bob", "eve", "tom"]
})

df2 = pd.DataFrame({
    "id" : [4, 5, 6],
    "name" : ["rahul", "tomba", "ram"]
})
```

```python
df1
```

```python
df2
```

```python
# row wise concatenation

pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
# column wise concatenation

pd.concat([df1, df2], axis=1)
```
