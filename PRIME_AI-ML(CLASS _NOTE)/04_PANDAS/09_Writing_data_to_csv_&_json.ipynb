```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df2 = df.copy()
```

```python
df2 = df2.drop_duplicates()
df2
```

```python
mean_age = df["age"].mean()

df2["age"] = df2["age"].fillna(mean_age)
df2
```

```python
mean_income = df["income"].mean()

df2["income"] = df2["income"].fillna(mean_income)
df2
```

```python
df2["gender"] = df2["gender"].ffill()
df2
```

```python
df2.dropna(inplace=True)
df2
```

```python
df2.sort_values("income", inplace=True)
df2
```

```python
df2.reset_index(drop=True, inplace=True)
df2
```

```python
# save file

df.to_csv("sorted_data.csv")
```
