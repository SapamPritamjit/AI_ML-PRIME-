### Melt :-
* {Wide to Long format (decreasing no of column and increasing no of row)}
* In wide format, the same kind of measurement is spread across multiple columns.
* In long format, those columns are stacked into two columns:
1) one for the variable name
2) one for the value

### Info :-
* id_vars :- Columns to keep fixed (identifiers).
* value_vars :- Columns to unpivot (turn into rows).
* var_name :- Name of the new column that stores former column names.
* value_name :- Name of the new column that stores values.
```python
df.melt(
    id_vars=["id"],
    value_vars=["math", "science"],
    var_name="subject",
    value_name="score"
)


````python
import pandas as pd
````

````python
df = pd.DataFrame({
    "country" : ["USA", "USA", "India", "India"],
    "year" : [2020, 2021, 2020, 2021],
    "sales" : [100, 120, 90, 110],
    "profit" : [20, 25, 18, 22],
})

df
````

````python
melt_df = df.melt(
    id_vars= ["country", "year"],
    value_vars= ["sales", "profit"],
    var_name= "metrics",
    value_name= "value"
)
melt_df
````

### Pivot :-
* (long to wide format)
* Takes one column’s unique values
* Spreads them out into multiple columns
* Fills those columns with values from another column
* It’s essentially the reverse of .melt().


### info :-
* index :- Column(s) to use as the row index (the identifier).
* columns :- Column whose unique values become new column names.
* values :- Column whose values fill the table.
```python
df.pivot(
    index = ["countru", "year"], # the column that want to keep
    columns = "metrics", 
    values = "value"
)

````python
pivot_df = melt_df.pivot(
    index = ["country", "year"],
    columns = "metrics",
    values= "value"
)
pivot_df
````
