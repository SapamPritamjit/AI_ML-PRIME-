* GroupBy :- It is used to split data into groups based on some columns, then apply a function like sum(), mean(), count(), etc. & combine the result

```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data.csv")
df
```

```python
df.groupby("country")["income"].mean()
```

```python
df.groupby("country")["income"].max()
```

```python
df.groupby("country")["income"].min()
```

```python
df.groupby("gender")["income"].mean()
```

### Aggregation :-
* It help us to use statical (summery) function to groupby data
* It help us to use multiple function at one time

```python
df.groupby("gender")["income"].agg(["min", "max", "mean"])
```

```python
df["income"].agg(["min", "max", "mean"])
```

* We can assign the column name too while using aggrigation function

```python
df.groupby("country")["income"].agg(min_salary="min", max_salary = "max", avg_Salary="mean")
```

```python
df.groupby("country").agg({
    "income" : "max",
    "age" : "mean"
})
```

```python
# label name change

df.groupby("country").agg(
    max_salary = ("income", "max"),
    avg_age = ("age", "mean")
)
```
