### label VS Index
* index mean position
* label mean named index

```python
# series -> 1D laballed array

s = pd.Series([15, 28, 32, 44, 51])
print(s)

# index
print(s[0])
```

```python
s = pd.Series([15, 28, 32, 44, 51], index=["Tomba", "Bob", "Charlie", "Chaoba", "Tom"])
s
```

```python
# excess using label name
print(s["Tom"])
print(s["Bob"])

# excess using index
print(s[0])
```

## Series Properties
* Homogenous
* Vectorized Operations
* Handle Missing value with NaN
* Mutable values, immutable size

```python
# Homogenous mean store same data type

s = pd.Series([15, 28, 32, 44, 51, "abc"], index=["Tomba", "Bob", "Charlie", "Chaoba", "Tom", 23])
print(s)
```

```python
# Vectorized operation

s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([6, 7, 8, 9, 10])
print(s1+s2)
```

```python
# Mutable values means we can modify

print(s1)
s1[0] = 22
print(s1)
```

```python
# immutable size means we cannot change the size if we do it will create a new series
print(s1)

change_s1 = s1.drop(0)
print(change_s1)
```
