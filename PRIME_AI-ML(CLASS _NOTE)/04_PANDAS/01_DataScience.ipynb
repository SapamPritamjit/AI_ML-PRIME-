## Q. What is EDA?

--> EDA is the process of examining and understanding a dataset before doing formal modeling or hypothesis testing. Its goal is to uncover patterns, spot anomalies, test assumptions, and check data quality using summary statistics and visualizations.

### Why EDA matters
* Helps you understand what the data is really saying
* Reveals missing values, outliers, and errors
* Guides feature selection and modeling choices
* Prevents wrong conclusions from poor data quality

### Common EDA activities
1. Data overview
* Dataset size (rows, columns)
* Data types (numeric, categorical, dates)
* Missing values

2. Summary statistics
* Mean, median, min, max
* Standard deviation
* Value counts for categories

3. Visualizations
* Histograms (distribution)
* Box plots (outliers)
* Scatter plots (relationships)
* Correlation matrices

4. Pattern & anomaly detection
* Trends
* Clusters
* Outliers or unexpected values
