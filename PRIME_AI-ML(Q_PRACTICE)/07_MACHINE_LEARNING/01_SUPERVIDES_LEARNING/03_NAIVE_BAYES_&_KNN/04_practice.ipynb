```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_classification(
    n_samples=2000,
    n_features=21,
    n_informative=18,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=0.2
)
```

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_lr.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
pipeline = Pipeline([
    ("lr", LogisticRegression())
])

param_grid = [
    {
        "lr__solver": ["liblinear"],
        "lr__penalty": ["l1", "l2"],
        "lr__C": [0.1, 0.5, 1, 5, 10],
        "lr__class_weight": ["balanced"],
        "lr__max_iter": [100, 300, 500]
    },
    {
        "lr__solver": ["saga"],
        "lr__penalty": ["l1", "l2", "elasticnet"],
        "lr__l1_ratio": [0.2, 0.5, 0.8],  # only used for elasticnet
        "lr__C": [0.1, 0.5, 1, 5, 10],
        "lr__class_weight": ["balanced"],
        "lr__max_iter": [100, 300, 500]
    },
    {
        "lr__solver": ["lbfgs"],
        "lr__penalty": ["l2"],
        "lr__C": [0.1, 0.5, 1, 5, 10],
        "lr__class_weight": ["balanced"],
        "lr__max_iter": [100, 300, 500]
    }
]

model_lr_cv = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    cv = 5,
    n_jobs=-1
)

model_lr_cv.fit(X_train, y_train)
y_pred = model_lr_cv.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_lr_cv.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```
