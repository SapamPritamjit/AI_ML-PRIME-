```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
X, y = make_classification(
    n_samples=2000,
    n_features=21,
    n_informative=18,
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, test_size=0.2
)
```

```python
from sklearn.linear_model import LogisticRegression

model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_lr.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
pipeline = Pipeline([
    ("lr", LogisticRegression())
])

param_grid = [
    {
        "lr__solver": ["liblinear"],
        "lr__penalty": ["l1", "l2"],
        "lr__C": [0.1, 0.5, 1, 5, 10],
        "lr__class_weight": ["balanced"],
        "lr__max_iter": [100, 300, 500]
    },
    {
        "lr__solver": ["saga"],
        "lr__penalty": ["l1", "l2", "elasticnet"],
        "lr__l1_ratio": [0.2, 0.5, 0.8],  # only used for elasticnet
        "lr__C": [0.1, 0.5, 1, 5, 10],
        "lr__class_weight": ["balanced"],
        "lr__max_iter": [100, 300, 500]
    },
    {
        "lr__solver": ["lbfgs"],
        "lr__penalty": ["l2"],
        "lr__C": [0.1, 0.5, 1, 5, 10],
        "lr__class_weight": ["balanced"],
        "lr__max_iter": [100, 300, 500]
    }
]

model_lr_cv = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    cv = 5,
    n_jobs=-1
)

model_lr_cv.fit(X_train, y_train)
y_pred = model_lr_cv.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_lr_cv.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier

scaler = StandardScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)

model_knn = KNeighborsClassifier()
model_knn.fit(X_train_scale, y_train)
y_pred = model_knn.predict(X_test_scale)

print("Training accuracy: ",accuracy_score(y_train, model_knn.predict(X_train_scale)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("knn", KNeighborsClassifier())
])

param_grid = [
    {
    "knn__n_neighbors" : [3, 5, 7, 9, 11, 13, 15],
    "knn__weights" : ["uniform", "distance"],
    "knn__algorithm" : ["auto", "brute"],
    "knn__p" : [1, 2]
    },
    {
    "knn__n_neighbors" : [3, 5, 7, 9, 11, 13, 15],
    "knn__weights" : ["uniform", "distance"],
    "knn__algorithm" : ["ball_tree", "kd_tree"],
    "knn__p" : [1, 2]
    }
]

model_knn_cv = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    cv = 5,
    n_jobs=-1
)

model_knn_cv.fit(X_train, y_train)
y_pred = model_knn_cv.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_knn_cv.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

model_nb = GaussianNB()
model_nb.fit(X_train, y_train)
y_pred = model_nb.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_nb.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
model_nb_cv = GridSearchCV(
    estimator=GaussianNB(),
    param_grid={"var_smoothing": [1e-12, 1e-11, 1e-10, 1e-9, 1e-8]},
    cv = 5,
    n_jobs=-1
)

model_nb_cv.fit(X_train, y_train)
y_pred = model_nb_cv.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_nb_cv.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier

model_dt = DecisionTreeClassifier()

model_dt.fit(X_train_scale, y_train)
y_pred = model_dt.predict(X_test_scale)

print("Training accuracy: ",accuracy_score(y_train, model_dt.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))
```

```python
pipeline = Pipeline([
    ("dt", DecisionTreeClassifier(random_state=42))
])

param_grid = {
    "dt__criterion": ["gini", "entropy"],
    "dt__max_depth": [5, 6, 7, 8, 9, 10],
    "dt__min_samples_split": [2, 5, 6, 7, 8, 9, 10],
    "dt__min_samples_leaf": [1, 2, 3, 4, 5],
    "dt__max_features": ["sqrt", "log2", None]
}

model_dt_cv = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    cv = 5,
    n_jobs=-1
)

model_dt_cv.fit(X_train, y_train)
y_pred = model_dt_cv.predict(X_test)

print("Training accuracy: ",accuracy_score(y_train, model_dt_cv.predict(X_train)))
print("Testing accuracy: ", accuracy_score(y_test, y_pred))
print("Recall: ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred))
print("confusion matric: \n", confusion_matrix(y_test, y_pred))

print("Best Parameters:", model_dt_cv.best_params_)
print("Best CV Score:", model_dt_cv.best_score_)
```
